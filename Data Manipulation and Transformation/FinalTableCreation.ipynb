```python
housingProblems_questions = df[df['QUESTION_ID'] == 8258]
requestForAssistance_questions = df[df['QUESTION_ID'] == 14208]
```

```python
housing_dummy_columns = housingProblems_questions['RESULTS'].str.get_dummies(';')
assistance_dummy_columns = requestForAssistance_questions['RESULTS'].str.get_dummies(';')
housing_id_column = housingProblems_questions['ID']
assistance_id_column = requestForAssistance_questions['ID']
```

```python
housing_df = pd.concat([housing_id_column, housing_dummy_columns], axis = 1)
assistance_df = pd.concat([assistance_id_column, assistance_dummy_columns], axis = 1)
```

```python
housing_df['Smoke detectors missing or not working'] = (
    housing_df['Smoke detectors missing or not working'] | housing_df['Smoke deterctors missing or not working'])
housing_df['None of the above'] = housing_df['None of the above'] | housing_df['Prefer not to answer']
housing_df.drop(columns = ['Prefer not to answer', 'Smoke deterctors missing or not working'], inplace = True)
```
