```python
import sqlite3
import pandas as pd
```

```python
conn = sqlite3.connect('SDOH.db')
main_table = 'Main'
main_query = f"SELECT * FROM {main_table} WHERE QUESTION_ID = '14208'"
```

```python
df = pd.read_sql_query(main_query, conn)
```

```python
df['RESULTS'][1]
```

```python
column_order = ['ID', 'Food', 'Transportation', 'Internet Access', 'Housing stability',
                'Housing problems (mold, pests, leaks, lead, smoke detectors)', 'Utilities (electric, gas, oil, water)',
                'Finances', 'Employment', 'Education', 'Stress', 'Social Connections', 'Physical Activity']
```

```python
dummy_columns = df['RESULTS'].str.get_dummies(';')
id_column = df['ID']
assistance_df = pd.concat([id_column, dummy_columns], axis = 1)
```

```python
set(column_order).symmetric_difference(set(dummy_columns.columns))
```

```python
column_order.extend(['NA', 'No assistance needed'])
```

```python
assistance_df = assistance_df.reindex(columns = column_order)
assistance_df
```

```python
assistance_df.to_sql('RequestForAssistance', conn, if_exists = 'replace', index = False)
```

```python
conn.commit()
conn.close()
```
